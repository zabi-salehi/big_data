```python
import pandas as pd
```

```python
data = pd.read_csv("netflix_titles.csv")
data.head()
```

```python
data = data.dropna()
data = data.drop(["type", "date_added", "rating"], axis=1)
data = data.rename({"listed_in":"genre"}, axis=1)
```

```python
data["director"] = data["director"].str.split(",").str[0]
```

```python
print(len(data))
data.head()
```

```python
# https://stackoverflow.com/a/51629081
def SQL_INSERT_STATEMENT_FROM_DATAFRAME(SOURCE, TARGET):
    sql_texts = []
    for index, row in SOURCE.iterrows():       
        sql_texts.append('INSERT INTO '+TARGET+' ('+ str(', '.join(SOURCE.columns))+ ') VALUES '+ str(tuple(row.values)) + ';')        
    return sql_texts
```

```python
target = "netflix_titles"
results = SQL_INSERT_STATEMENT_FROM_DATAFRAME(data, target)
result = '\n'.join(results)
with open('mariadb-dump.sql', 'w', encoding="utf-8") as f:
    f.write(result)
```
